```python
import matplotlib.pyplot as plt 
import numpy as np
x = np.linspace(-10, 10, 100)
y = np.sin(x)
xnum = np.linspace(-10, 10, 11)
ynum =np.linspace(-1, 1, 11)

plt.plot(x, y, color="r", label="sin θ")
#lt.ylabel("Magnitude")
#lt.xlabel("Angel θ in Radians")
plt.xticks(xnum)
plt.yticks(ynum)
plt.grid()
plt.legend()
plt.title("Plot of $ f(θ)=\sin θ $")
plt.axes().spines["bottom"].set_position(('data', 0))
plt.axes().spines["left"].set_position(('data', 0))
plt.axhline(y=plt.ylim()[0], color='k')
plt.axvline(x=-10.1, color='k')
plt.axis([-10.1, 10.1, -1.1, 1.1])
#lt.annotate(s='sine local maxima', xy=(0.5*np.pi, np.sin(0.5*np.pi)), xytext=(1.0*np.pi, 0.75))
plt.show()
```
